```python
# If needed in Colab:
# !pip install -q transformers datasets accelerate sentencepiece

import os
import json
import random
import shutil
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import torch

from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import (
    accuracy_score,
    precision_recall_fscore_support,
    confusion_matrix,
    classification_report,
    roc_auc_score,
    roc_curve
)

from transformers import (
    AutoTokenizer,
    AutoModelForSequenceClassification,
    TrainingArguments,
    Trainer,
    DataCollatorWithPadding,
    EarlyStoppingCallback,
    set_seed
)
from datasets import Dataset


# =========================================================
# 0) SETTINGS
# =========================================================
RANDOM_SEED = 42
random.seed(RANDOM_SEED)
np.random.seed(RANDOM_SEED)
torch.manual_seed(RANDOM_SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(RANDOM_SEED)
set_seed(RANDOM_SEED)

TRAIN_CSV = "/content/train.csv"
VAL_CSV = "/content/validation.csv"
TEST_CSV = "/content/test.csv"

TITLE_COL = "title"
TEXT_COL = "full_text"
LABEL_COL = "class_label"

MODEL_NAME = "lvbert_5fold_cv"
LVBERT_MODEL_NAME = "AiLab-IMCS-UL/lvbert"

ARTIFACT_DIR = f"/content/artifacts/{MODEL_NAME}"
os.makedirs(ARTIFACT_DIR, exist_ok=True)

MODEL_DIR = os.path.join(ARTIFACT_DIR, "final_model")
TOKENIZER_DIR = os.path.join(ARTIFACT_DIR, "final_tokenizer")

CV_RESULTS_PATH = os.path.join(ARTIFACT_DIR, "cv_fold_results.csv")
CV_TUNING_PATH = os.path.join(ARTIFACT_DIR, "cv_all_tuning_results.csv")
CV_SUMMARY_PATH = os.path.join(ARTIFACT_DIR, "cv_summary_statistics.csv")
CV_PARAM_SUMMARY_PATH = os.path.join(ARTIFACT_DIR, "cv_param_summary.csv")

FINAL_PREDICTIONS_PATH = os.path.join(ARTIFACT_DIR, "final_test_predictions.csv")
FINAL_METRICS_JSON_PATH = os.path.join(ARTIFACT_DIR, "final_test_metrics.json")
FINAL_REPORT_PATH = os.path.join(ARTIFACT_DIR, "final_test_classification_report.txt")
FINAL_CONFUSION_MATRIX_PATH = os.path.join(ARTIFACT_DIR, "final_test_confusion_matrix.png")
FINAL_ROC_PATH = os.path.join(ARTIFACT_DIR, "final_test_roc_curve.png")
FINAL_TRAINING_HISTORY_PATH = os.path.join(ARTIFACT_DIR, "final_training_history.csv")

SEMANTIC_DIR = os.path.join(ARTIFACT_DIR, "for_semantic_signal_building")
os.makedirs(SEMANTIC_DIR, exist_ok=True)

SEMANTIC_TRAIN_PRED_PATH = os.path.join(SEMANTIC_DIR, "lvbert_train_predictions.csv")
SEMANTIC_VAL_PRED_PATH = os.path.join(SEMANTIC_DIR, "lvbert_validation_predictions.csv")
SEMANTIC_TEST_PRED_PATH = os.path.join(SEMANTIC_DIR, "lvbert_test_predictions.csv")

MAX_LEN_GRID = [256]
LEARNING_RATE_GRID = [2e-5, 3e-5]
BATCH_SIZE_GRID = [8]

NUM_EPOCHS = 3
WEIGHT_DECAY = 0.01
EARLY_STOPPING_PATIENCE = 1
N_FOLDS = 5

print("Artifact directory ready:", ARTIFACT_DIR)


# =========================================================
# 1) HELPER FUNCTIONS
# =========================================================
def reset_seeds(seed=RANDOM_SEED):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)
    set_seed(seed)


def clean_directory(path):
    if os.path.exists(path):
        shutil.rmtree(path)
    os.makedirs(path, exist_ok=True)


def load_df(path: str):
    df = pd.read_csv(path, encoding="utf-8")

    for col in [TITLE_COL, TEXT_COL, LABEL_COL]:
        if col not in df.columns:
            raise ValueError(f"Missing column: {col} in {path}")

    df[TITLE_COL] = df[TITLE_COL].fillna("").astype(str)
    df[TEXT_COL] = df[TEXT_COL].fillna("").astype(str)
    df[LABEL_COL] = df[LABEL_COL].astype(int)
    return df


def get_prediction_group(true_label, pred_label):
    if true_label == 1 and pred_label == 1:
        return "TP"
    elif true_label == 0 and pred_label == 0:
        return "TN"
    elif true_label == 0 and pred_label == 1:
        return "FP"
    elif true_label == 1 and pred_label == 0:
        return "FN"
    return "UNKNOWN"


def softmax_numpy(logits):
    logits = np.asarray(logits)
    logits = logits - np.max(logits, axis=1, keepdims=True)
    exp_vals = np.exp(logits)
    return exp_vals / np.sum(exp_vals, axis=1, keepdims=True)


def evaluate_model(y_true, y_pred, y_score):
    acc = accuracy_score(y_true, y_pred)
    p, r, f1, _ = precision_recall_fscore_support(
        y_true,
        y_pred,
        average="binary",
        zero_division=0
    )
    cm = confusion_matrix(y_true, y_pred)
    roc_auc = roc_auc_score(y_true, y_score)
    report_text = classification_report(y_true, y_pred, digits=4, zero_division=0)

    metrics_dict = {
        "accuracy": float(acc),
        "precision": float(p),
        "recall": float(r),
        "f1_score": float(f1),
        "roc_auc": float(roc_auc)
    }

    metrics_df = pd.DataFrame({
        "metric": ["accuracy", "precision", "recall", "f1_score", "roc_auc"],
        "value": [acc, p, r, f1, roc_auc]
    })

    return metrics_dict, metrics_df, cm, report_text


def plot_confusion_matrix(cm, save_path):
    plt.figure(figsize=(6, 5))
    sns.heatmap(
        cm,
        annot=True,
        fmt="d",
        cmap="Blues",
        xticklabels=["Class 0", "Class 1"],
        yticklabels=["Class 0", "Class 1"],
        cbar=False,
        square=True,
        linewidths=0.5
    )
    plt.xlabel("Predicted label")
    plt.ylabel("True label")
    plt.title("Final test confusion matrix")
    plt.tight_layout()
    plt.savefig(save_path, dpi=300, bbox_inches="tight")
    plt.show()


def plot_roc_curve(y_true, y_score, roc_auc, save_path):
    fpr, tpr, _ = roc_curve(y_true, y_score)

    plt.figure(figsize=(6, 5))
    plt.plot(fpr, tpr, label=f"AUC = {roc_auc:.4f}", linewidth=2)
    plt.plot([0, 1], [0, 1], "--", linewidth=1)
    plt.xlabel("False positive rate")
    plt.ylabel("True positive rate")
    plt.title("Final test ROC-AUC curve")
    plt.legend(loc="lower right")
    plt.grid(True, linestyle="--", alpha=0.4)
    plt.tight_layout()
    plt.savefig(save_path, dpi=300, bbox_inches="tight")
    plt.show()


def save_prediction_results(df_input, y_true, y_pred, y_prob, output_path):
    results_df = df_input[[TITLE_COL, TEXT_COL, LABEL_COL]].copy()
    results_df["pred_label"] = y_pred
    results_df["pred_prob"] = y_prob
    results_df["text_length"] = results_df[TEXT_COL].fillna("").astype(str).str.len()

    results_df["group"] = [
        get_prediction_group(true_label, pred_label)
        for true_label, pred_label in zip(y_true, y_pred)
    ]

    results_df = results_df[
        [
            TITLE_COL,
            TEXT_COL,
            LABEL_COL,
            "pred_label",
            "pred_prob",
            "group",
            "text_length"
        ]
    ]

    results_df.to_csv(output_path, index=False, encoding="utf-8")
    return results_df


def make_hf_dataset(df):
    hf_df = pd.DataFrame({
        "title": df[TITLE_COL].fillna("").astype(str),
        "text": df[TEXT_COL].fillna("").astype(str),
        "label": df[LABEL_COL].astype(int)
    })
    return Dataset.from_pandas(hf_df, preserve_index=False)


def tokenize_batch(batch, tokenizer, max_len):
    return tokenizer(
        batch["title"],
        batch["text"],
        truncation=True,
        padding=False,
        max_length=max_len
    )


def compute_metrics_trainer(eval_pred):
    logits, labels = eval_pred
    probs = softmax_numpy(logits)[:, 1]
    preds = np.argmax(logits, axis=1)

    acc = accuracy_score(labels, preds)
    p, r, f1, _ = precision_recall_fscore_support(
        labels,
        preds,
        average="binary",
        zero_division=0
    )
    roc_auc = roc_auc_score(labels, probs)

    return {
        "accuracy": acc,
        "precision": p,
        "recall": r,
        "f1": f1,
        "roc_auc": roc_auc
    }


def prepare_dataset_for_model(df, tokenizer, max_len):
    ds = make_hf_dataset(df)
    ds = ds.map(
        lambda batch: tokenize_batch(batch, tokenizer, max_len),
        batched=True
    )
    return ds


def predict_with_trainer(trainer, dataset):
    pred_output = trainer.predict(dataset)
    logits = pred_output.predictions
    probs = softmax_numpy(logits)[:, 1]
    preds = np.argmax(logits, axis=1)
    return preds, probs, logits


# =========================================================
# 2) LOAD DATA
# =========================================================
df_train = load_df(TRAIN_CSV)
df_val = load_df(VAL_CSV)
df_test = load_df(TEST_CSV)

df_trainval = pd.concat([df_train, df_val], ignore_index=True)

print("\nLoaded datasets:")
print(f"Train: {len(df_train)}")
print(f"Validation: {len(df_val)}")
print(f"Train+Validation: {len(df_trainval)}")
print(f"Test: {len(df_test)}")

print("\nClass distribution:")
print("Train:", df_train[LABEL_COL].value_counts().to_dict())
print("Validation:", df_val[LABEL_COL].value_counts().to_dict())
print("Train+Validation:", df_trainval[LABEL_COL].value_counts().to_dict())
print("Test:", df_test[LABEL_COL].value_counts().to_dict())

X_trainval_y = df_trainval[LABEL_COL].values


# =========================================================
# 3) 5-FOLD CROSS-VALIDATION ON TRAIN+VALIDATION
# =========================================================
skf = StratifiedKFold(
    n_splits=N_FOLDS,
    shuffle=True,
    random_state=RANDOM_SEED
)

cv_fold_rows = []
cv_tuning_rows = []

for fold_idx, (train_idx, val_idx) in enumerate(skf.split(df_trainval, X_trainval_y), start=1):
    print("\n" + "=" * 70)
    print(f"STARTING FOLD {fold_idx}/{N_FOLDS}")

    fold_train_df = df_trainval.iloc[train_idx].reset_index(drop=True)
    fold_val_df = df_trainval.iloc[val_idx].reset_index(drop=True)

    best_row = None

    for max_len in MAX_LEN_GRID:
        for learning_rate in LEARNING_RATE_GRID:
            for batch_size in BATCH_SIZE_GRID:
                reset_seeds()

                trial_output_dir = os.path.join(
                    ARTIFACT_DIR,
                    f"fold_{fold_idx}_trial_len{max_len}_lr{learning_rate}_bs{batch_size}"
                )
                clean_directory(trial_output_dir)

                tokenizer = AutoTokenizer.from_pretrained(LVBERT_MODEL_NAME)
                model = AutoModelForSequenceClassification.from_pretrained(
                    LVBERT_MODEL_NAME,
                    num_labels=2
                )

                train_ds = prepare_dataset_for_model(fold_train_df, tokenizer, max_len)
                val_ds = prepare_dataset_for_model(fold_val_df, tokenizer, max_len)

                data_collator = DataCollatorWithPadding(tokenizer=tokenizer)

                training_args = TrainingArguments(
                    output_dir=trial_output_dir,
                    eval_strategy="epoch",
                    save_strategy="epoch",
                    logging_strategy="epoch",
                    learning_rate=learning_rate,
                    per_device_train_batch_size=batch_size,
                    per_device_eval_batch_size=batch_size,
                    num_train_epochs=NUM_EPOCHS,
                    weight_decay=WEIGHT_DECAY,
                    load_best_model_at_end=True,
                    metric_for_best_model="f1",
                    greater_is_better=True,
                    save_total_limit=1,
                    seed=RANDOM_SEED,
                    report_to="none",
                    disable_tqdm=False
                )

                trainer = Trainer(
                    model=model,
                    args=training_args,
                    train_dataset=train_ds,
                    eval_dataset=val_ds,
                    processing_class=tokenizer,
                    data_collator=data_collator,
                    compute_metrics=compute_metrics_trainer,
                    callbacks=[EarlyStoppingCallback(early_stopping_patience=EARLY_STOPPING_PATIENCE)]
                )

                trainer.train()
                eval_result = trainer.evaluate()

                log_history = trainer.state.log_history
                trained_epochs = sum(
                    1 for row in log_history
                    if ("loss" in row and "epoch" in row and "eval_loss" not in row)
                )

                row = {
                    "fold": fold_idx,
                    "max_len": max_len,
                    "learning_rate": learning_rate,
                    "batch_size": batch_size,
                    "val_accuracy": eval_result["eval_accuracy"],
                    "val_precision": eval_result["eval_precision"],
                    "val_recall": eval_result["eval_recall"],
                    "val_f1": eval_result["eval_f1"],
                    "val_roc_auc": eval_result["eval_roc_auc"],
                    "epochs_trained": trained_epochs
                }
                cv_tuning_rows.append(row)

                if best_row is None or row["val_f1"] > best_row["val_f1"]:
                    best_row = row

                del trainer
                del model
                if torch.cuda.is_available():
                    torch.cuda.empty_cache()

    print("Best row for this fold:")
    print(best_row)

    reset_seeds()

    fold_best_output_dir = os.path.join(ARTIFACT_DIR, f"fold_{fold_idx}_best_model")
    clean_directory(fold_best_output_dir)

    tokenizer = AutoTokenizer.from_pretrained(LVBERT_MODEL_NAME)
    model = AutoModelForSequenceClassification.from_pretrained(
        LVBERT_MODEL_NAME,
        num_labels=2
    )

    train_ds = prepare_dataset_for_model(fold_train_df, tokenizer, int(best_row["max_len"]))
    val_ds = prepare_dataset_for_model(fold_val_df, tokenizer, int(best_row["max_len"]))
    data_collator = DataCollatorWithPadding(tokenizer=tokenizer)

    training_args = TrainingArguments(
        output_dir=fold_best_output_dir,
        eval_strategy="epoch",
        save_strategy="epoch",
        logging_strategy="epoch",
        learning_rate=float(best_row["learning_rate"]),
        per_device_train_batch_size=int(best_row["batch_size"]),
        per_device_eval_batch_size=int(best_row["batch_size"]),
        num_train_epochs=NUM_EPOCHS,
        weight_decay=WEIGHT_DECAY,
        load_best_model_at_end=True,
        metric_for_best_model="f1",
        greater_is_better=True,
        save_total_limit=1,
        seed=RANDOM_SEED,
        report_to="none",
        disable_tqdm=False
    )

    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=train_ds,
        eval_dataset=val_ds,
        processing_class=tokenizer,
        data_collator=data_collator,
        compute_metrics=compute_metrics_trainer,
        callbacks=[EarlyStoppingCallback(early_stopping_patience=EARLY_STOPPING_PATIENCE)]
    )

    trainer.train()

    fold_preds, fold_probs, _ = predict_with_trainer(trainer, val_ds)
    y_fold_val = fold_val_df[LABEL_COL].astype(int).values

    fold_metrics, _, _, _ = evaluate_model(
        y_true=y_fold_val,
        y_pred=fold_preds,
        y_score=fold_probs
    )

    log_history = trainer.state.log_history
    trained_epochs = sum(
        1 for row in log_history
        if ("loss" in row and "epoch" in row and "eval_loss" not in row)
    )

    fold_row = {
        "fold": fold_idx,
        "best_max_len": int(best_row["max_len"]),
        "best_learning_rate": float(best_row["learning_rate"]),
        "best_batch_size": int(best_row["batch_size"]),
        "train_size": len(fold_train_df),
        "val_size": len(fold_val_df),
        "epochs_trained": trained_epochs,
        **fold_metrics
    }
    cv_fold_rows.append(fold_row)

    del trainer
    del model
    if torch.cuda.is_available():
        torch.cuda.empty_cache()

cv_results_df = pd.DataFrame(cv_fold_rows)
cv_results_df.to_csv(CV_RESULTS_PATH, index=False, encoding="utf-8")

cv_tuning_df = pd.DataFrame(cv_tuning_rows)
cv_tuning_df.to_csv(CV_TUNING_PATH, index=False, encoding="utf-8")

cv_summary_df = pd.DataFrame([{
    "num_folds": N_FOLDS,
    "accuracy_mean": cv_results_df["accuracy"].mean(),
    "accuracy_std": cv_results_df["accuracy"].std(ddof=1),
    "precision_mean": cv_results_df["precision"].mean(),
    "precision_std": cv_results_df["precision"].std(ddof=1),
    "recall_mean": cv_results_df["recall"].mean(),
    "recall_std": cv_results_df["recall"].std(ddof=1),
    "f1_mean": cv_results_df["f1_score"].mean(),
    "f1_std": cv_results_df["f1_score"].std(ddof=1),
    "roc_auc_mean": cv_results_df["roc_auc"].mean(),
    "roc_auc_std": cv_results_df["roc_auc"].std(ddof=1)
}])
cv_summary_df.to_csv(CV_SUMMARY_PATH, index=False, encoding="utf-8")

print("\n=== 5-FOLD CV RESULTS ===")
print(cv_results_df.sort_values("f1_score", ascending=False).to_string(index=False))

print("\n=== 5-FOLD CV SUMMARY ===")
print(cv_summary_df.to_string(index=False))


# =========================================================
# 4) PARAMETER SUMMARY AND FINAL HYPERPARAMETERS
# =========================================================
param_summary = (
    cv_tuning_df
    .groupby(["max_len", "learning_rate", "batch_size"], as_index=False)
    .agg(
        val_accuracy_mean=("val_accuracy", "mean"),
        val_precision_mean=("val_precision", "mean"),
        val_recall_mean=("val_recall", "mean"),
        val_f1_mean=("val_f1", "mean"),
        val_roc_auc_mean=("val_roc_auc", "mean"),
        epochs_trained_mean=("epochs_trained", "mean")
    )
    .sort_values(["val_f1_mean", "val_roc_auc_mean"], ascending=[False, False])
)

param_summary.to_csv(CV_PARAM_SUMMARY_PATH, index=False, encoding="utf-8")

best_final_row = param_summary.iloc[0]
final_max_len = int(best_final_row["max_len"])
final_learning_rate = float(best_final_row["learning_rate"])
final_batch_size = int(best_final_row["batch_size"])

print("\n=== PARAMETER SUMMARY ACROSS FOLDS ===")
print(param_summary.to_string(index=False))

print("\nChosen final hyperparameters:")
print({
    "max_len": final_max_len,
    "learning_rate": final_learning_rate,
    "batch_size": final_batch_size
})


# =========================================================
# 5) FINAL TRAINING ON ALL TRAIN+VALIDATION
# =========================================================
clean_directory(MODEL_DIR)
clean_directory(TOKENIZER_DIR)

reset_seeds()

tokenizer = AutoTokenizer.from_pretrained(LVBERT_MODEL_NAME)
trainval_ds = prepare_dataset_for_model(df_trainval, tokenizer, final_max_len)

data_collator = DataCollatorWithPadding(tokenizer=tokenizer)

model = AutoModelForSequenceClassification.from_pretrained(
    LVBERT_MODEL_NAME,
    num_labels=2
)

training_args = TrainingArguments(
    output_dir=MODEL_DIR,
    eval_strategy="no",
    save_strategy="epoch",
    logging_strategy="epoch",
    learning_rate=final_learning_rate,
    per_device_train_batch_size=final_batch_size,
    per_device_eval_batch_size=final_batch_size,
    num_train_epochs=NUM_EPOCHS,
    weight_decay=WEIGHT_DECAY,
    load_best_model_at_end=False,
    save_total_limit=1,
    seed=RANDOM_SEED,
    report_to="none",
    disable_tqdm=False
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=trainval_ds,
    processing_class=tokenizer,
    data_collator=data_collator
)

trainer.train()

trainer.save_model(MODEL_DIR)
tokenizer.save_pretrained(TOKENIZER_DIR)

history_df = pd.DataFrame(trainer.state.log_history)
history_df.to_csv(FINAL_TRAINING_HISTORY_PATH, index=False, encoding="utf-8")

best_model = AutoModelForSequenceClassification.from_pretrained(MODEL_DIR)
best_tokenizer = AutoTokenizer.from_pretrained(TOKENIZER_DIR)

predict_trainer = Trainer(
    model=best_model,
    processing_class=best_tokenizer,
    data_collator=DataCollatorWithPadding(tokenizer=best_tokenizer)
)

train_pred_ds = prepare_dataset_for_model(df_train, best_tokenizer, final_max_len)
val_pred_ds = prepare_dataset_for_model(df_val, best_tokenizer, final_max_len)
test_pred_ds = prepare_dataset_for_model(df_test, best_tokenizer, final_max_len)

train_pred, train_prob, _ = predict_with_trainer(predict_trainer, train_pred_ds)
val_pred, val_prob, _ = predict_with_trainer(predict_trainer, val_pred_ds)
test_pred, test_prob, _ = predict_with_trainer(predict_trainer, test_pred_ds)

y_train = df_train[LABEL_COL].astype(int).values
y_val = df_val[LABEL_COL].astype(int).values
y_test = df_test[LABEL_COL].astype(int).values


# =========================================================
# 6) FINAL TEST EVALUATION
# =========================================================
metrics_dict, metrics_df, cm, report_text = evaluate_model(
    y_true=y_test,
    y_pred=test_pred,
    y_score=test_prob
)

metrics_dict["final_max_len"] = int(final_max_len)
metrics_dict["final_learning_rate"] = float(final_learning_rate)
metrics_dict["final_batch_size"] = int(final_batch_size)
metrics_dict["num_epochs"] = int(NUM_EPOCHS)
metrics_dict["weight_decay"] = float(WEIGHT_DECAY)
metrics_dict["pretrained_checkpoint"] = LVBERT_MODEL_NAME

print("\n=== FINAL TEST RESULTS ===")
print(metrics_df.to_string(index=False))

print("\nConfusion matrix [[TN, FP], [FN, TP]]:")
print(cm)

print("\nClassification report:")
print(report_text)


# =========================================================
# 7) SAVE FINAL METRICS + REPORT
# =========================================================
with open(FINAL_METRICS_JSON_PATH, "w", encoding="utf-8") as f:
    json.dump(metrics_dict, f, ensure_ascii=False, indent=2)

with open(FINAL_REPORT_PATH, "w", encoding="utf-8") as f:
    f.write(report_text)

print("\nSaved final files:")
print(f"Model dir: {MODEL_DIR}")
print(f"Tokenizer dir: {TOKENIZER_DIR}")
print(f"Final metrics JSON: {FINAL_METRICS_JSON_PATH}")
print(f"Final classification report: {FINAL_REPORT_PATH}")
print(f"Final training history: {FINAL_TRAINING_HISTORY_PATH}")


# =========================================================
# 8) SAVE PREDICTION CSVs
# =========================================================
final_test_results_df = save_prediction_results(
    df_input=df_test,
    y_true=y_test,
    y_pred=test_pred,
    y_prob=test_prob,
    output_path=FINAL_PREDICTIONS_PATH
)

save_prediction_results(
    df_input=df_train,
    y_true=y_train,
    y_pred=train_pred,
    y_prob=train_prob,
    output_path=SEMANTIC_TRAIN_PRED_PATH
)

save_prediction_results(
    df_input=df_val,
    y_true=y_val,
    y_pred=val_pred,
    y_prob=val_prob,
    output_path=SEMANTIC_VAL_PRED_PATH
)

save_prediction_results(
    df_input=df_test,
    y_true=y_test,
    y_pred=test_pred,
    y_prob=test_prob,
    output_path=SEMANTIC_TEST_PRED_PATH
)

print(f"\nFinal prediction file saved: {FINAL_PREDICTIONS_PATH}")
print(f"Semantic train predictions saved: {SEMANTIC_TRAIN_PRED_PATH}")
print(f"Semantic validation predictions saved: {SEMANTIC_VAL_PRED_PATH}")
print(f"Semantic test predictions saved: {SEMANTIC_TEST_PRED_PATH}")

print("\nPrediction file preview:")
print(final_test_results_df.head().to_string(index=False))


# =========================================================
# 9) PLOT + SAVE FINAL CONFUSION MATRIX
# =========================================================
plot_confusion_matrix(cm, FINAL_CONFUSION_MATRIX_PATH)
print(f"Final confusion matrix saved: {FINAL_CONFUSION_MATRIX_PATH}")


# =========================================================
# 10) PLOT + SAVE FINAL ROC CURVE
# =========================================================
plot_roc_curve(
    y_true=y_test,
    y_score=test_prob,
    roc_auc=metrics_dict["roc_auc"],
    save_path=FINAL_ROC_PATH
)
print(f"Final ROC curve saved: {FINAL_ROC_PATH}")


# =========================================================
# 11) FINAL MESSAGE
# =========================================================
print("\nLVBERT 5-fold CV experiment finished successfully.")
print("You now have:")
print("- 5-fold CV fold results")
print("- all CV tuning results")
print("- CV summary statistics")
print("- parameter summary across folds")
print("- final trained model directory")
print("- final tokenizer directory")
print("- final test metrics JSON")
print("- final classification report")
print("- final test prediction CSV")
print("- final training history CSV")
print("- final confusion matrix PNG")
print("- final ROC curve PNG")
print("- for_semantic_signal_building folder")
print("  - lvbert_train_predictions.csv")
print("  - lvbert_validation_predictions.csv")
print("  - lvbert_test_predictions.csv")
```

```python
import pandas as pd
import numpy as np
from scipy.stats import gaussian_kde

PREDICTIONS_FILE = "/content/artifacts/lvbert_5fold_cv/for_semantic_signal_building/lvbert_test_predictions.csv"
OUTPUT_PATH = "/content/lvbert_kde_analysis_stats.csv"

df = pd.read_csv(PREDICTIONS_FILE)

order = ["TP", "TN", "FP", "FN"]

results = []

x_range = np.linspace(0, 1, 1000)

for group in order:
    values = df[df["group"] == group]["pred_prob"].dropna()

    if len(values) > 1:
        kde = gaussian_kde(values)
        y = kde(x_range)

        peak_x = x_range[np.argmax(y)]
        peak_y = np.max(y)

        results.append({
            "group": group,
            "count": len(values),
            "mean": values.mean(),
            "median": values.median(),
            "std": values.std(),
            "min": values.min(),
            "max": values.max(),
            "peak_probability": peak_x,
            "peak_density": peak_y
        })

results_df = pd.DataFrame(results)
results_df.to_csv(OUTPUT_PATH, index=False)

print(results_df)
print("\nSaved:", OUTPUT_PATH)
```

*   **Maximum Sequence Length (MAX_LEN_GRID):** `[256]`

_The final selected hyperparameters after cross-validation are determined dynamically based on `val_f1_mean` and `val_roc_auc_mean`._

*   **Random Seed:** `42`
*   **Number of Epochs (NUM_EPOCHS):** `3`
*   **Weight Decay (WEIGHT_DECAY):** `0.01`

*   **Number of Folds for Cross-Validation (N_FOLDS):** `5`
*   **Evaluation Strategy:** `epoch` (during CV and final training, with no evaluation for final training after CV to preserve computation)
*   **Save Strategy:** `epoch`
*   **Logging Strategy:** `epoch`
*   **Metric for Best Model:** `f1` (used during cross-validation for early stopping and model loading)
*   **Greater is Better:** `True` (for the metric `f1`)
*   **Save Total Limit:** `1` (keeps only the best model checkpoint)
*   **Report to:** `none` (no external reporting for Hugging Face Trainer)
